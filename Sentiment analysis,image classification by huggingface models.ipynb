```python
import transformers
from transformers import pipeline
pipe= pipeline("text-classification", model = "distilbert-base-uncased-finetuned-sst-2-english")
pipe("This movie is just mehhh")
```

```python
nlp= pipeline("ner")
example="My name is maria and i am going to Pakistan"

ner_results = nlp(example)
print(ner_results)
```

```python
nlp= pipeline("ner",model = "dslim/bert-base-NER")
example="My name is asad and i am going to Pakistan"

ner_results = nlp(example)
print(ner_results)
```

```python
import transformers
from transformers  import pipeline
pipe= pipeline("question-answering")
question="what is the name of capital city of Pakistan?"
context="the name of capital city of Pakistan is Islamabad."
pipe(question=question,context=context)
```
